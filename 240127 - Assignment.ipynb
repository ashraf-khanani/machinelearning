```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv('canada_per_capita_income.csv')

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.sample(15)
```

```python
df = df.rename(columns={'per capita income (US$)': 'income'})

df.info()
```

```python
plt.xlabel('income')
plt.ylabel('year')
plt.scatter(df.income,df.year,color='red',marker='.')
```

```python
new_df = df.drop('income',axis='columns')

new_df
```

```python
income = df.income

income
```

```python
reg = linear_model.LinearRegression()

reg.fit(new_df,income)
```

```python
reg.predict([[2017]])
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
year_df = pd.read_csv('years.csv')

year_df.head(5)
```

```python
work = reg.predict(year_df)

work
```

```python
year_df['income'] = work
year_df
```

```python
plt.xlabel('Year',fontsize =20)
plt.ylabel('Income',fontsize =20)
plt.scatter(df.year,df.income,color="red" ,marker=".")
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
#plt.scatter(df.income,df.year,color='green',marker='.')
```

```python
year_df.to_csv("incomepredicted.csv")
```
